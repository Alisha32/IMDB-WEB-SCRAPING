```python
#IMPORTING RELEVANT LIBRARIES
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
#LOADING THE WEBPAGE
r=requests.get('https://www.imdb.com/chart/top/')
#converting into beautiful soup object
soup = bs(r.content)
contents = soup.prettify()
print(contents[:100])
```

```python
#CREATING EMPTY LISTS
movie_title=[]
movie_year=[]
movie_rating=[]
```

```python
#EXTRACTING HTML TAAGS
imdb_table = soup.find(class_="chart full-width")
movie_titlecolumn = imdb_table.find_all(class_="titleColumn")
movie_ratingscolumn = imdb_table.find_all(class_="ratingColumn imdbRating")
```

```python
#CREATING LIST
for row in movie_titlecolumn:
  title = row.a.text
  movie_title.append(title)
movie_title
```

```python
for row in movie_titlecolumn:
  year=row.span.text
  movie_year.append(year)
movie_year
```

```python
for row in movie_ratingscolumn:
  rating=row.strong.text
  movie_rating.append(rating)
movie_rating
```

```python
#CREATING DATAFRAME
movie_df = pd.DataFrame({ 'Movie Title':movie_title, 'Year Of Release':movie_year, 'Rating': movie_rating})
movie_df.head(30)
```
